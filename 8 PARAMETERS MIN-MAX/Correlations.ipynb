```python
from tabulate import tabulate

selected_columns =  ['Weigh_Feeder_rate_ A(TPH)', 'Weigh_feeder_rate_B(TPH)', 'Ball_mill_amps',
            'BALL_MILL_VENT_SYS_TEMP(Â°C)','BALL_MILL_VENT_FAN_AMPS', 'BAG_FILTER_DP(mmWC)',
            'SCREW_CONV _ CURRENT(Amps)','BUCKET_ELEVATOR_CURRENT(Amps)']


corr_matrix = data[selected_columns].corr()

print(tabulate(corr_matrix, headers='keys', tablefmt='fancy_grid'))

# Plot the heatmap
# plt.figure(figsize=(20, 10))  # adjust the figsize as necessary
# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
# plt.title('Selected Feature Correlation Matrix')
# plt.show()
```

```python
# Define the index range for the columns to be accessed (5th to 42nd columns)
start_index = 5
end_index = 42

# Get the column names for the specified range
features = data.columns[start_index:end_index]

corr_matrix = data[features].corr()

print(tabulate(corr_matrix, headers='keys', tablefmt='fancy_grid'))

corr_matrix.to_excel('correlation matrix(process parameters).xlsx')
```

```python
target_vars = ['> 35','35 to 200', '< 200']
feature_vars = [col for col in features if col not in target_vars]

for target in target_vars:
    target_corr = data[feature_vars].corrwith(data[target])
    print(f'Correlation with {target}:')
    print(target_corr, '\n')
```
